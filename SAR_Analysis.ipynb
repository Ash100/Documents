```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your CSV
df = pd.read_csv('/content/S24-BI162-.csv')  # Replace with the actual filename
```

```python
# Set parameter as index for better plotting
df.set_index('Parameter', inplace=True)
```

```python
# Convert percentage strings to float if needed
df = df.astype(float)
```

```python
# === Plot 2: Line Chart of Overall Satisfaction ===
fig, ax = plt.subplots(figsize=(10, 6))  # Use fig, ax for saving

df['Overall Rating/Satisfaction Level(%)'].plot(marker='o', linestyle='-', ax=ax)

ax.set_title('Overall Satisfaction by Parameter')
ax.set_ylabel('Overall Satisfaction (%)')
ax.set_xticks(ticks=range(len(df.index)))
ax.set_xticklabels(df.index, rotation=90)
ax.grid(True)

# Save figure at 600 DPI before showing
fig.savefig('S24-BI162-.png', dpi=600, bbox_inches='tight')

plt.tight_layout()
plt.show()
```

```python
# === Plot 1: Stacked Bar Chart ===
df_rating = df[['Excellent(%)', 'Very Good(%)', 'Good(%)', 'Average(%)', 'Poor(%)']]

fig, ax = plt.subplots(figsize=(12, 12))  # Wider figure to accommodate legend
df_rating.plot(kind='barh', stacked=True, colormap='viridis', ax=ax)

ax.set_title('Student Ratings by Parameter')
ax.set_xlabel('Percentage')

# Place the legend outside without shrinking the plot
ax.legend(title='Rating', bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0.)

# Manually adjust layout to make space for the legend
plt.subplots_adjust(right=0.78)

# Save figure at 600 DPI
fig.savefig('S24-BI162.png', dpi=600, bbox_inches='tight')

plt.show()
```
